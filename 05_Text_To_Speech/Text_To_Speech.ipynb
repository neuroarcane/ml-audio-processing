# Text To Speech
## This notebook outlines the concepts invovled in Converting a piece of text into speech

Google Text to Speech API - **gTTS**

### Features

- Customizable speech-specific sentence tokenizer that allows for unlimited lengths of text to be read, all while keeping proper intonation, abbreviations, decimals

- Customizable text pre-processors which can, for example, provide pronunciation corrections

- Automatic retrieval of supported languages

### Steps
- 1. Install the library
- 2. Import the library
- 3. Create Text-to-Speech object
- 4. Create a text
- 5. Send the text for conversion

### Import the library

```python
from gtts import gTTS
import IPython.display as ipd
import os
```

### Sample text to convert

```python
with open('sample_text.txt', 'r') as f:
    text = f.read()
print(text[:300] + '...')
```

### Specify language

```python
language = 'en'
print(f"Language set to: {language}")
```

### Convert Text to Speech

```python
tts = gTTS(text=text, lang=language, slow=False)
print("Text-to-Speech object created successfully")
```

### Save the converted speech

```python
tts.save('output_speech_en.mp3')
print("Speech saved to output_speech_en.mp3")
```

### Play the speech

```python
import IPython.display
IPython.display.display(IPython.display.Audio('output_speech_en.mp3'))
```

### Try some other text

```python
other_text = ("Natural language processing is a subfield of artificial intelligence "
              "that focuses on the interaction between computers and human language, "
              "enabling machines to read, decipher, understand, and make sense of human languages.")
tts2 = gTTS(text=other_text, lang='en', slow=False)
print("Second TTS object created")
```

## Other Language Conversions - Hindi

## Convert from Text files

```python
with open('sample_text.txt', 'r') as f:
    file_text = f.read()
print(f"File loaded: {len(file_text)} characters")
print("Preview:", file_text[:200])
```

```python
# Convert first 1500 characters to avoid very long audio
tts_file = gTTS(text=file_text[:1500], lang='en', slow=False)
print("TTS object from file text created")
```

```python
tts_file.save('output_from_text_file.mp3')
print("Saved to output_from_text_file.mp3")
```

```python
import IPython.display
IPython.display.display(IPython.display.Audio('output_from_text_file.mp3'))
```

```python
print("Generated audio files:")
for f in sorted(os.listdir('.')):
    if f.endswith('.mp3'):
        size_kb = os.path.getsize(f) / 1024
        print(f"  {f}  ({size_kb:.1f} KB)")
```
